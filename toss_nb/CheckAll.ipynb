```python
import pandas as pd

sig1 = []
sig2 = []
ratio_dict = {}

for k,v in fitting_info_dict.items():
    if v["mean_1"] >= v["mean_2"]:
        right = v["sigma_1"]/v["mean_1"]
        sig2.append(right)
        left = v["sigma_2"]/v["mean_2"]
        sig1.append(left) 
    else:
        right = v["sigma_2"]/v["mean_2"]
        sig2.append(right)
        left = v["sigma_1"]/v["mean_1"]
        sig1.append(left)
    ratio_dict[k] = {"right":right, "left":left}
sig3 = sig1+sig2
        
ps1 = pd.value_counts(list(map(lambda x:round(x,3), sig1)))
ps1 = sorted([(i,j) for i,j in zip(ps1.index, ps1.values)], key = lambda x:x[0])
x1 = [xy[0] for xy in ps1]
y1 = [xy[1] for xy in ps1]

ps2 = pd.value_counts(list(map(lambda x:round(x,3), sig2)))
ps2 = sorted([(i,j) for i,j in zip(ps2.index, ps2.values)], key = lambda x:x[0])
x2 = [xy[0] for xy in ps2]
y2 = [xy[1] for xy in ps2]

ps3 = pd.value_counts(list(map(lambda x:round(x,3), sig3)))
ps3 = sorted([(i,j) for i,j in zip(ps3.index, ps3.values)], key = lambda x:x[0])
x3 = [xy[0] for xy in ps3]
y3 = [xy[1] for xy in ps3]
```

```python
def gauss_plot(info):
    (num,mean,sigma) = info
    sigma = 0.001 if sigma == 0 else sigma
    fig = plt.figure(figsize=(4,4))
    ax1 = plt.subplot(111)
    #fig = plt.figure(figsize=(4,4))
    x_lim = [mean-4*sigma-0.5, mean+4*sigma+0.5]
    y_lim = [-0.05,1.05]
    points = (x_lim[1]-x_lim[0])/0.001
    x = np.linspace(x_lim[0], x_lim[1], int(points))
    y = list(map(lambda x:np.exp((-(x-mean)**2)/(2*(sigma**2))),x))
    plt.yticks([])
    ax1.plot(x,y)
    ax1.text(s="mean=%s sigma=%s"%(mean,sigma), y=1.05, x=mean, size=8,ha="center")
    ax1.vlines(ymin=0,ymax=1.05,x=mean,linestyles="dashed",color="r")
    return fig
```
